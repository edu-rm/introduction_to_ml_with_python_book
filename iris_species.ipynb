```
# CONHECENDO OS DADOS

from sklearn.datasets import load_iris
iris_dataset = load_iris()

# print(format(iris_dataset.keys()))
# print(format(iris_dataset['data']))
# print("\n{}".format(iris_dataset['DESCR']))
# print("\n{}".format(iris_dataset['target_names']))

# 150 samples and 4 features
print(iris_dataset['data'].shape)
```

```
# TREINANDO E TESTANDO DADOS

# X maíusculo pois X é um array de duas dimensões
# y (target) minúsculo pois y possui uma dimensão

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0
)

print(X_train.shape)
print(y_train.shape)

```

```
# Antes de criar um modelo precisamos inspencionar os dados, para ver se preci - 
# samos mesmo de ML, pra ver se o que eu desejo está nos dados

# É interessante pra descobrir anormalidades e peculiaridades
# EX: se estão todas na mesma unidade

# Melhor maneira: visualizando.
#   - Scatter plot. 
#     - 1 feature no eixo X e outra no eixo Y
#   - Pair plot
#     - Todos os pares possíveis de features

# Para plotar: Converter NumPy array em DataFrame
#   - Podemos plotar par de feature com scatter_matrix

import matplotlib.pyplot as plt
import pandas as pd

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

grr = pd.plotting.scatter_matrix(
      iris_dataframe, 
      c=y_train, 
      figsize=(16,16), 
      marker='o',
      hist_kwds={'bins': 20}, 
      s=60, 
      alpha=.8
      # cmap=mglearn.cm3   
)

# As classes estão bem separadas significando que um modelo de ML terá a capacidade
# de aprender a separar elas 
```
